```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

### Dataset Number 1

```python
print("Titanic Dataset Exploration")
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)
```

```python
print(titanic.isnull().sum()) #checking the missing values for data exploration
```

```python
print(titanic.describe()) #basic summary stat
```

```python
survival_rate = titanic.groupby('Pclass')['Survived'].mean()
print("Survival Rate by Passenger Class:\n", survival_rate)
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic['Age'].dropna(), kde=True, bins=30, color='blue')
plt.title("Distribution of Passenger Ages", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Pclass", y="Survived", data=titanic, ci=None, palette="muted")
plt.title("Survival Rate by Passenger Class", fontsize=16)
plt.xlabel("Passenger Class", fontsize=12)
plt.ylabel("Survival Rate", fontsize=12)
plt.xticks([0, 1, 2], ["1st Class", "2nd Class", "3rd Class"])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Sex", y="Survived", data=titanic, ci=None, palette="Set2")
plt.title("Survival Rate by Gender", fontsize=16)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Survival Rate", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="Survived", y="Age", data=titanic, palette="coolwarm")
plt.title("Age Distribution by Survival Status", fontsize=16)
plt.xlabel("Survived (0 = No, 1 = Yes)", fontsize=12)
plt.ylabel("Age", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x="Embarked", hue="Survived", data=titanic, palette="pastel")
plt.title("Survival Count by Embarkation Port", fontsize=16)
plt.xlabel("Port of Embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

## Dataset Number 2

```python
iris = load_iris()
iris_data = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_data['species'] = iris.target
```

```python
import plotly.express as px
import pandas as pd

species_names = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
iris_data['species'] = iris_data['species'].map(species_names)

fig = px.scatter(
    iris_data,
    x='sepal length (cm)',
    y='sepal width (cm)',
    size='petal length (cm)',
    color='species',
    title='Interactive Iris Dataset Scatter Plot',
    labels={'sepal length (cm)': 'Sepal Length (cm)', 'sepal width (cm)': 'Sepal Width (cm)', 'petal length (cm)': 'Petal Length (cm)', 'species': 'Species'},
    hover_data=['petal width (cm)']
)

fig.show()
```
